```python
nsm_13 = pd.read_csv('nsm_semidaily_df_2013.csv')
nlg_13 = pd.read_csv('nlg_semidaily_df_2013.csv')
nsmz_13 = pd.read_csv('nsmz_semidaily_df_2013.csv')
nmdz_13 = pd.read_csv('nmdz_semidaily_df_2013.csv')
nlgz_13 = pd.read_csv('nlgz_semidaily_df_2013.csv')
nsm_14 = pd.read_csv('nsm_semidaily_df_2014.csv')
nlg_14 = pd.read_csv('nlg_semidaily_df_2014.csv')
nsmz_14 = pd.read_csv('nsmz_semidaily_df_2014.csv')
nmdz_14 = pd.read_csv('nmdz_semidaily_df_2014.csv')
nlgz_14 = pd.read_csv('nlgz_semidaily_df_2014.csv')
no3_13 = pd.read_csv('no3_semidaily_df_2013.csv')
po4_13 = pd.read_csv('po4_semidaily_df_2013.csv')
sio4_13 = pd.read_csv('sio4_semidaily_df_2013.csv')
no3_14 = pd.read_csv('no3_semidaily_df_2014.csv')
po4_14 = pd.read_csv('po4_semidaily_df_2014.csv')
sio4_14 = pd.read_csv('sio4_semidaily_df_2014.csv')
```

```python
nsm_13.head()
```

```python
# For 2013 and 2014 semidaily csvs, take avg every 10076 rows, create and append dates vector, then relabel, then merge
```

```python
nsm_13=nsm_13.groupby(np.arange(len(nsm_13))//10076).mean()
nlg_13=nlg_13.groupby(np.arange(len(nlg_13))//10076).mean()
nsmz_13=nsmz_13.groupby(np.arange(len(nsmz_13))//10076).mean()
nmdz_13=nmdz_13.groupby(np.arange(len(nmdz_13))//10076).mean()
nlgz_13=nlgz_13.groupby(np.arange(len(nlgz_13))//10076).mean()
no3_13=no3_13.groupby(np.arange(len(no3_13))//10076).mean()
po4_13=po4_13.groupby(np.arange(len(po4_13))//10076).mean()
sio4_13=sio4_13.groupby(np.arange(len(sio4_13))//10076).mean()

nsm_14=nsm_14.groupby(np.arange(len(nsm_14))//10076).mean()
nlg_14=nlg_14.groupby(np.arange(len(nlg_14))//10076).mean()
nsmz_14=nsmz_14.groupby(np.arange(len(nsmz_14))//10076).mean()
nmdz_14=nmdz_14.groupby(np.arange(len(nmdz_14))//10076).mean()
nlgz_14=nlgz_14.groupby(np.arange(len(nlgz_14))//10076).mean()
no3_14=no3_14.groupby(np.arange(len(no3_14))//10076).mean()
po4_14=po4_14.groupby(np.arange(len(po4_14))//10076).mean()
sio4_14=sio4_14.groupby(np.arange(len(sio4_14))//10076).mean()
```

```python
dates_13 = pd.DataFrame([])
dates_13["Time"] = pd.date_range(start="2013-01-02 12:00:00",
                                   end = '2013-12-28 00:00:00',
                                   freq = '12H' )

dates_14 = pd.DataFrame([])
dates_14["Time"] = pd.date_range(start="2014-01-02 12:00:00",
                                   end = '2014-12-28 00:00:00',
                                   freq = '12H' )
```

```python
df1 = pd.concat([dates_13, nsm_13], axis=1)
df1=df1.rename(columns={"concentration": "nsm"})

df1 = pd.concat([df1, nlg_13], axis=1)
df1=df1.rename(columns={"concentration": "nlg"})

df1 = pd.concat([df1, nsmz_13], axis=1)
df1=df1.rename(columns={"concentration": "nsmz"})

df1 = pd.concat([df1, nmdz_13], axis=1)
df1=df1.rename(columns={"concentration": "nmdz"})

df1 = pd.concat([df1, nlgz_13], axis=1)
df1=df1.rename(columns={"concentration": "nlgz"})

df1 = pd.concat([df1, no3_13], axis=1)
df1=df1.rename(columns={"concentration": "no3"})

df1 = pd.concat([df1, sio4_13], axis=1)
df1=df1.rename(columns={"concentration": "sio4"})

df1 = pd.concat([df1, po4_13], axis=1)
df1=df1.rename(columns={"concentration": "po4"})

del df1['year']
del df1['Unnamed: 0']
df1.head()
```

```python
df2 = pd.concat([dates_14, nsm_14], axis=1)
df2=df2.rename(columns={"concentration": "nsm"})

df2 = pd.concat([df2, nlg_14], axis=1)
df2=df2.rename(columns={"concentration": "nlg"})

df2 = pd.concat([df2, nsmz_14], axis=1)
df2=df2.rename(columns={"concentration": "nsmz"})

df2 = pd.concat([df2, nmdz_14], axis=1)
df2=df2.rename(columns={"concentration": "nmdz"})

df2 = pd.concat([df2, nlgz_14], axis=1)
df2=df2.rename(columns={"concentration": "nlgz"})

df2 = pd.concat([df2, no3_14], axis=1)
df2=df2.rename(columns={"concentration": "no3"})

df2 = pd.concat([df2, sio4_14], axis=1)
df2=df2.rename(columns={"concentration": "sio4"})

df2 = pd.concat([df2, po4_14], axis=1)
df2=df2.rename(columns={"concentration": "po4"})

del df2['year']
del df2['Unnamed: 0']
df2.head()
```

```python
frames = [df1, df2]
semidaily_means = pd.concat(frames, axis=0)
semidaily_means.head()
```

```python
pd.DataFrame.to_csv(semidaily_means, 'semidaily_means.csv')
```
